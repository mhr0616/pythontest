```python
# 代码3-41
import pandas as pd
import numpy as np
musicdata = pd.read_csv('./data/musicdata.csv',dtype={'value_actual':np.float64})
musicdataGroup = musicdata[['format',  'value_actual']].groupby(by='format')
print('分组后的音乐行业收入信息表为:', list(musicdataGroup))
```

```python
company=["A","B","C"]
data=pd.DataFrame({
    "company":[company[x] for x in np.random.randint(0,len(company),10)],
    "salary":np.random.randint(5,50,10),
    "age":np.random.randint(15,50,10)
}
)
print(data)
```

```python
group = data.groupby("company")
print(list(group))
```

针对样例数据集，如果我想求不同公司员工的平均年龄和平均薪水，可以按照下方的代码进行：

```python
data.groupby("company").agg('mean')
```

如果想对针对不同的列求不同的值，比如要计算不同公司员工的平均年龄以及薪水的中位数，可以利用字典进行聚合操作的指定：

```python
data.groupby('company').agg({'salary':'median','age':'mean'})
```
