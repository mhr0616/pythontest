2.绘制y=x^2曲线和添加y=x^4曲线的曲线

```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline表示在行中显示图片，在命令行运行报错
data = np.arange(0, 1.1, 0.01)
plt.title('lines')  # 添加标题
plt.xlabel('x')  # 添加x轴的名称
plt.ylabel('y')  # 添加y轴的名称
plt.xlim((0, 1))  # 确定x轴范围
plt.ylim((0, 1))  # 确定y轴范围
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 规定x轴刻度
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 确定y轴刻度
plt.plot(data, data ** 2)  # 添加y=x^2曲线
plt.plot(data, data ** 4)  # 添加y=x^4曲线
plt.legend(['y=x^2', 'y=x^4'])
plt.savefig('./tmp/y=x^2.jpg')
plt.show()
```

p1 = plt.figure(figsize=(8, 6), dpi=80)  # 确定画布大小

这行代码使用matplotlib库创建一个新的画布，并指定了画布的大小和分辨率。具体来说，figsize参数指定了画布的宽度和高度（以英寸为单位），dpi参数指定了画布的分辨率（每英寸点数）。在这里，画布的大小为8英寸（宽）×6英寸（高），分辨率为80点/英寸。

创建画布后，可以在画布上绘制图形，例如使用plot函数绘制折线图、使用scatter函数绘制散点图等。最后，使用show函数显示画布中的图形

```python
# 代码5-2
#这行代码使用numpy库的arange函数生成一个从0到2π之间，步长为0.01的数组，并将其保存在变量x中。
#arange函数的参数分别是起始值、终止值和步长。在这里，起始值是0，终止值是2π（即一个圆的周长），步长是0.01。
# 所以x数组中的元素是从0开始，每次增加0.01，直到接近2π但不包括2π。

x = np.arange(0, np.pi * 2, 0.01)
# 第一幅子图
p1 = plt.figure(figsize=(8, 6), dpi=80)  # 确定画布大小
# 创建一个2行1列的子图，并开始绘制第一幅
ax1 = p1.add_subplot(2, 1, 1)
plt.title('lines')  # 添加标题
plt.xlabel('x')  # 添加x轴的名称
plt.ylabel('y')  # 添加y轴的名称
plt.xlim((0, 1))  # 确定x轴范围
plt.ylim((0, 1))  # 确定y轴范围
plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 规定x轴刻度
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1])  # 确定y轴刻度
plt.plot(x, x ** 2)  # 添加y=x^2曲线
plt.plot(x, x ** 4)  # 添加y=x^4曲线
plt.legend(['y=x^2', 'y=x^4'])

ax2 = p1.add_subplot(2, 1, 2)  # 开始绘制第二幅
plt.title('sin/cos(x)')  # 添加标题
plt.xlabel('x')  # 添加x轴的名称
plt.ylabel('y')  # 添加y轴的名称
plt.xlim((0, np.pi * 2))  # 确定x轴范围
plt.ylim((-1, 1))  # 确定y轴范围
plt.xticks([0, np.pi / 2, np.pi, np.pi * 1.5, np.pi * 2])  # 规定x轴刻度
plt.yticks([-1, -0.5, 0, 0.5, 1])  # 确定y轴刻度
plt.plot(x, np.sin(x))  # 添加sin(x)曲线
plt.plot(x, np.cos(x))  # 添加cos(x)曲线
plt.legend(['y=sin(x)', 'y=cos(x)'])
plt.tight_layout()  # 调整两个子图间距
plt.savefig('./tmp/sincos(x).jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2*np.pi, 0.1)
y1 = np.tan(x)
y2 = 1 / np.tan(x)   # tan(0)的倒数是无穷大  y2 = np.where(np.isclose(y1, 0), 1e10, 1 / y1)

plt.plot(x, y1, label='tan(x)')
plt.plot(x, y2, label='cot(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of tan(x) and cot(x)')
plt.legend()
plt.show()
```

### 2. 设置pyplot的动态rc参数(PPT P7)


```python
# 代码5-3
# 原图
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi)  # 生成x轴数据
y = np.sin(x)  # 生成y轴数据
plt.plot(x, y, label='$sin(x)$')  # 绘制sin曲线图
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### （1）长相问题

```python
# 修改rc参数后的图
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.plot(x, y, label='$sin(x)$')  # 绘制三角函数
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
